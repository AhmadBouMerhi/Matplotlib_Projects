```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#set value to an axis for monthly temparature 
x_axis_data = np.arange(1,13,1)
x_axis_data
```

```python
#plot monthly averages temperature in New Jersey
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
points_C = [(x-32) * 0.56 for x in points]
```

```python
# Plot both C and F
fahrenheit, = plt.plot(x_axis_data, points, marker="+",color="blue", linewidth=1, label="Fahreneit")
celcius, = plt.plot(x_axis_data, points_C, marker="s", color="Red", linewidth=1, label="Celcius")
```

```python
#Assign title to the chart 
plt.legend(handles=[fahrenheit, celcius], loc="best")
```

```python
# Create labels for the X and Y axis
plt.xlabel("Months")
plt.ylabel("Degrees")
```

```python
# Save and display the chart
plt.savefig("../Images/avg_temp.png")
plt.show()
```
